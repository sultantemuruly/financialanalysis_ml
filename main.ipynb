```python
path = '/kaggle/input/financial-statements-of-major-companies2009-2023/Financial Statements.csv'
df = pd.read_csv(path)
df
```

```python
df = df.rename(columns={'Company ': 'Company'})
df.columns.tolist()
```

```python
object_cols = df.select_dtypes(include='object').columns
int_cols = df.select_dtypes(include='int').columns
float_cols = df.select_dtypes(include='float').columns

print("Columns with dtype object: ", object_cols.tolist(), 'count: ', len(object_cols.tolist()))
print("Columns with dtype int: ", int_cols.tolist(), 'count: ', len(int_cols.tolist()))
print("Columns with dtype float: ", float_cols.tolist(), 'count: ', len(float_cols.tolist()))
print(len(df.columns.tolist()))
```

```python
print('Year unique: ', df['Year'].nunique())
print('Company unique: ', df['Company'].nunique())
print('Category unique: ', df['Category'].nunique())
```

```python
from sklearn.model_selection import train_test_split

df_cleaned = df.dropna(subset=['Market Cap(in B USD)'])

y = df_cleaned['Market Cap(in B USD)']
X = df_cleaned.drop(['Market Cap(in B USD)'], axis=1)

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)

categorical_cols = df.select_dtypes(include='object').columns.tolist()
```

**Using pipeline**

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

model = RandomForestRegressor(n_estimators=100, random_state=0)

my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                             ])

my_pipeline.fit(X_train, y_train)
predictions = my_pipeline.predict(X_valid)
score = mean_absolute_error(y_valid, predictions)
print('MAE:', score)
```

**Cross-Validation**

```python
from sklearn.model_selection import cross_val_score

scores = -1 * cross_val_score(my_pipeline, X, y,
                              cv=5,
                              scoring='neg_mean_absolute_error')

print("MAE scores:\n", scores)
print("MAE average:\n", scores.mean())
```
